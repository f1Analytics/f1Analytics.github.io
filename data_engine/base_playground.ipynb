```python
import fastf1 as ff1

import pandas as pd
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship, sessionmaker
```

```python
# Define the base class for declarative
Base = declarative_base()

class QualiResult(Base):
    __tablename__ = 'quali_results'
    id = Column(Integer, primary_key=True)
    driverId = Column(String, ForeignKey('drivers.driverId'))
    teamId = Column(Integer, ForeignKey('constructor.constructorId'))
    raceId = Column(Integer, ForeignKey('races.raceId'))
    position = Column(Integer)
    q1 = Column(Integer)
    q2 = Column(Integer)
    q3 = Column(Integer)
    
    def __init__(self, driverId, teamId, raceId, position, q1, q2, q3):
        self.driverId = driverId
        self.teamId = teamId
        self.raceId = raceId
        self.position = position
        self.q1 = q1
        self.q2 = q2
        self.q3 = q3
        
    def __repr__(self):
        return f"QualiResult(driverId={self.driverId}, teamId={self.teamId}, raceId={self.raceId}, position={self.position}, q1={self.q1}, q2={self.q2}, q3={self.q3})"
    
class Driver(Base):
    __tablename__ = 'drivers'
    id = Column(Integer, primary_key=True)
    driverId = Column(String)
    abbreviation = Column(String)
    name = Column(String)
    middlename = Column(String)
    surname = Column(String)
    nationality = Column(String)
    
class Team(Base):
    __tablename__ = 'teams'
    id = Column(Integer, primary_key=True)
    team_name = Column(String)
    
    
# Define your SQLAlchemy models
class TelemetryData(Base):
    __tablename__ = 'telemetry_data'
    id = Column(Integer, primary_key=True)

class Races(Base):
    __tablename__ = 'races'
    id = Column(Integer, primary_key=True)
    countryId = driverId = Column(String, ForeignKey('countries.id'))
    
```

```python
engine = create_engine('sqlite:///testf1.db', echo=True)
Base.metadata.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()


```
